```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.svm import SVC

# создание набора данных полулун
x, y = make_moons(noise=0.1)

# создание и обучение модели SVM с линейным ядром
svm_linear = SVC(kernel='linear')
svm_linear.fit(x, y)

# создание и обучение модели SVM с RBF-ядром
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(x, y)

# создание сетки для визуализации
x1s = np.linspace(x[:, 0].min() - 1, x[:, 0].max() + 1, 100)
x2s = np.linspace(x[:, 1].min() - 1, x[:, 1].max() + 1, 100)
x1, x2 = np.meshgrid(x1s, x2s)
x_new = np.c_[x1.ravel(), x2.ravel()]

y_pred_linear = svm_linear.predict(x_new)
y_pred_linear = y_pred_linear.reshape(x1.shape)

y_pred_rbf = svm_rbf.predict(x_new)
y_pred_rbf = y_pred_rbf.reshape(x1.shape)

# визуализация графика линейной классификации
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(x[y == 0, 0], x[y == 0, 1], "bo")
plt.plot(x[y == 1, 0], x[y == 1, 1], "ro")
plt.contourf(x1, x2, y_pred_linear, alpha=0.3)
plt.title("линейная классификация")

# визуализация графика нелинейной классификации
plt.subplot(122)
plt.plot(x[y == 0, 0], x[y == 0, 1], "bo")
plt.plot(x[y == 1, 0], x[y == 1, 1], "ro")
plt.contourf(x1, x2, y_pred_rbf, alpha=0.3)
plt.title("нелинейная классификация")

plt.show()
```
