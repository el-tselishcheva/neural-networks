```python
import pandas as pd

# загрузка csv датасетов
train = pd.read_csv('./kaggle-data/train.csv')
train.shape
```

```python
# вывод первых строк данных из тренировочной выборки
train.head()
```

```python
# определение процентного значения отсутствующих данных в столбцах
def percent_missing(df):
    data = pd.DataFrame(df)
    df_cols = list(pd.DataFrame(data))
    dict_x = {}
    for i in range(0, len(df_cols)):
        dict_x.update({df_cols[i]: round(data[df_cols[i]].isnull().mean()*100,2)})
    
    return dict_x

missing = percent_missing(train)
df_miss = sorted(missing.items(), key=lambda x: x[1], reverse=True)
print('процент отсутствующих данных:')
df_miss
```

```python
# удаление столбцов, в которых отсутствует более 50% данных
train_new = train.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'MasVnrType'], axis=1)
train_new.shape
```

```python
# заполнение недостающих данных фиктивными значениями
def handle_missing(features):
    objects = []
    for i in features.columns:
        if features[i].dtype == object:
            objects.append(i)
    for o in objects:
        # замена NaN на наиболее часто встречающееся значение для строковых данных
        features[o] = features[o].fillna(features[o].mode().iloc[0])
        
    numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    numeric = []
    for i in features.columns:
        if features[i].dtype in numeric_dtypes:
            numeric.append(i)
    for n in numeric:
        # замена NaN на среднее значение для числовых данных
        features[n] = features[n].fillna(features[n].mean())

    return features

train_new = handle_missing(train_new)
train_new.head()
```

```python
# проверка, что все недостающие данные были заполнены
missing = percent_missing(train_new)
df_miss = sorted(missing.items(), key=lambda x: x[1], reverse=True)
print('процент отсутствующих данных (после заполнения):')
df_miss
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import r2_score

# замена всех категориальных переменных фиктивными
train_new = pd.get_dummies(train_new).reset_index(drop=True)

X = train_new.drop(['SalePrice', 'Id'], axis=1)
y = train_new['SalePrice']

# разделение данных на тренировочную и тестовую выборку в соотношении 8:2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# прогнозирование с помощью метода SVR
svr_model = SVR(kernel='linear')
svr_model.fit(X_train, y_train)
y_pred = svr_model.predict(X_test)

# оценка качества прогноза с помощью метрики R^2
r2 = r2_score(y_test, y_pred)
print(f'R^2 для SVR: {r2:f}')
```

```python
import numpy as np

preds = np.concatenate((y_train.values, y_pred))
preds[0:len(y_train)-1] = None
```

```python
import matplotlib.pyplot as plt

# сравнение реальных и спрогнозированных данных на графике
plt.figure()

plt.plot(train['Id'], train['SalePrice'], color='blue', label='фактическая цена')
plt.plot(train['Id'], preds, color='red', label='спрогнозированная цена')

plt.title('сравнение реальных и спрогнозированных цен')
plt.xlabel('порядковый номер')
plt.ylabel('цена продажи')
plt.legend()
plt.show()
```
